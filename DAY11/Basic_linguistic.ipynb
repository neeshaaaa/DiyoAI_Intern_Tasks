
# LINGUISTICS BASICS

Linguistics = Scientific study of language

Main branches useful in NLP:

1. Morphology
   -> Structure of words

   Example:
   "unhappiness"
   un + happy + ness

2. Syntax
   -> Sentence structure / grammar

   Example:
   "The cat sits on the mat."

3. Semantics
   -> Meaning of language

4. Pragmatics
   -> Meaning based on context

5. Phonology / Phonetics
   -> Sound systems
   Mostly important in speech processing
 
 

```python

# WHAT IS A TOKEN?

"""
Token = Smallest meaningful unit after splitting text.

Sentence:
"I love Natural Language Processing."

Possible tokens:
["I", "love", "Natural", "Language", "Processing", "."]

Tokenization is usually the FIRST step in NLP.
"""

text = "I love Natural Language Processing."

print(text)
```

# Word Tokenization with NLTK  

```python
from nltk.tokenize import word_tokenize

text = "I love Natural Language Processing."

tokens = word_tokenize(text)

print("Original Sentence:")
print(text)

print("\nTokens:")
print(tokens)
```

```python
# observing special characters

text = "Hello!!! NLP is fun :) #AI"

tokens = word_tokenize(text)

print(tokens)

# Notice:
# punctuation becomes separate tokens
# hashtags are preserved
# emoticons are handled separately
```

# Sentence Tokenization

```python
from nltk.tokenize import sent_tokenize

paragraph = """
Natural Language Processing is interesting.
It helps computers understand language.
Tokenization is the first important step."""

sentences = sent_tokenize(paragraph)

print("Sentence Tokenization:\n")

for i, sentence in enumerate(sentences):
    print(f"Sentence {i+1}: {sentence}")
```

Why do we tokenize text?

Computers cannot understand raw text directly.

Machines process:
- tokens
- numbers
- vectors

Tokenization helps:
1. Search engines
2. Chatbots
3. Translation systems
4. Speech assistants
5. Text summarization
6. Sentiment analysis

# Stopwords 

Stopwords are very common words:
- the
- is
- are
- and
- in

Sometimes removed to reduce noise.

```python
from nltk.corpus import stopwords

stop_words = stopwords.words('english')

print("Number of stopwords:", len(stop_words))

print("\nSome stopwords:")
print(stop_words[:30])
```

```python
# removing stopwords from a sentence

text = "This is a simple example showing stopword removal in NLP. I love NLP!. NLpI is really interesting."

tokens = word_tokenize(text)

filtered_tokens = []

for word in tokens:
    if word.lower() not in stop_words:
        filtered_tokens.append(word)

print("Original Tokens:")
print(tokens)

print("\nAfter Stopword Removal:")
print(filtered_tokens)
```

# Stemming : Cuts words mechanically.

Example:
- running -> run
- studies -> studi

Can produce non-real words.

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["running", "runner", "runs", "easily", "studies", "happily", "tolerance"]

print("Stemming Results:\n")

for word in words:
    stem = stemmer.stem(word)
    print(f"{word} ---> {stem}")
```

# Lemmatization: Converts words to dictionary base form.

More intelligent than stemming.



```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ["running", "better", "studies", "cars"]

print("Lemmatization Results:\n")

for word in words:
    lemma = lemmatizer.lemmatize(word)
    print(f"{word} ---> {lemma}")
```

# Better lemmatization

```python
# Verb-aware lemmatization

words = ["running", "eating", "driving"]

for word in words:
    lemma = lemmatizer.lemmatize(word, pos='v')
    print(f"{word} ---> {lemma}")

# pos='v' means verb
```

# POS Tagging

POS = Part of Speech

Examples:
1. NN  -> noun
2. VB  -> verb
3. JJ  -> adjective
4. RB  -> adverb

```python
from nltk import pos_tag

sentence = "The quick brown fox jumps over the lazy dog."

tokens = word_tokenize(sentence)

pos_tags = pos_tag(tokens)

print(pos_tags)
```

```python
# understanding POS tags

sentence = "My name is Neesha Basnet. I am pursuing my Bachelor's degree in BEIT"

tokens = word_tokenize(sentence)

tags = pos_tag(tokens)

for word, tag in tags:
    print(f"{word:15} ---> {tag}")
```

# Named Entity Recognition (NER)

NER identifies:
- people
- organizations
- countries
- cities
etc.

```python
from nltk import ne_chunk

sentence = "Rododendron is a national flower of Nepal."

tokens = word_tokenize(sentence)

tags = pos_tag(tokens)

named_entities = ne_chunk(tags)

print(named_entities)
```

# Introduction to spaCy

spaCy is a Python library used to process and analyze text efficiently for natural language processing tasks. It provides ready-to-use models and tools for working with linguistic data.

- Supports tokenization, POS tagging and dependency parsing
- Designed for speed and production use
- Works well with large text datasets
- Commonly used in NLP pipelines


Advantages:
- faster
- cleaner API
- industrial usage
- powerful pipelines

# load spaCy  model

```python
import spacy

# Load English language model
nlp = spacy.load("en_core_web_sm")

print("spaCy model loaded successfully!")
```

```python
#spaCy Tokenization

text = "I am learning NLP using spaCy."

doc = nlp(text)

print("Tokens:\n")

for token in doc:
    print(token.text)
```

```python
# token attributes in spaCy

text = "Riya bought a new iPhone."

doc = nlp(text)

for token in doc:
    print(f"""
TEXT      : {token.text}
LOWERCASE : {token.lower_}
IS_ALPHA  : {token.is_alpha}
IS_STOP   : {token.is_stop}
SHAPE     : {token.shape_}
""")
```

# Sentence detection in spaCy

```python
text = """
NLP is amazing.
spaCy makes NLP easier.
Tokenization is important.
"""

doc = nlp(text)

print("Sentences:\n")

for sent in doc.sents:
    print(sent)
```

# Pos tagging in spaCy

```python
text = "Today, my task is to learn about basics of linguistics and NLTK and spaCy."

doc = nlp(text)

for token in doc:
    print(f"""
WORD       : {token.text}
POS        : {token.pos_}
TAG        : {token.tag_}
EXPLANATION: {spacy.explain(token.tag_)}
""")
```

# lemmatization in spaCy

```python
text = "My laptop is too heavy."

doc = nlp(text)

for token in doc:
    print(f"{token.text:15} ---> {token.lemma_}")
```

# Named Entity Recognition in spaCy

```python
text = " I am doing NLP and AI internship in DiyoAI."

doc = nlp(text)

print("Named Entities:\n")

for ent in doc.ents:
    print(f"""
TEXT  : {ent.text}
LABEL : {ent.label_}
MEANING : {spacy.explain(ent.label_)}
""")
```

```python
# Comparing NLTK and spaCy

comparison = {
    "NLTK": [
        "Educational",
        "Good for learning",
        "More manual",
        "Research focused"
    ],
    
    "spaCy": [
        "Production ready",
        "Faster",
        "Cleaner API",
        "Industry focused"
    ]
}

for library, points in comparison.items():
    print(f"\n{library}")
    print("-" * 20)
    
    for point in points:
        print(point)
```
