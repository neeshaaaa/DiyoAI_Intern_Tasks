# Word embeddings : Word2Vec theory + Gensim hands-on

**Word embeddings** are techniques in Natural Language Processing (NLP) that represent words as dense, numeric vectors. They translate human language into coordinates within a multidimensional space, enabling machine learning models to grasp semantic meanings, syntactic relationships, and context.


# Why Do We Need Word Embeddings?

Consider the words:

- King
- Queen
- Prince
- Princess

Humans know that these words are related.

However, traditional methods such as One-Hot Encoding treat every word as completely independent.

As a result:
- King ≠ Queen
- King ≠ Prince

Word embeddings solve this problem by placing similar words close together in a vector space.

Words with similar meanings tend to have similar vector representations.

#### Importance:
- Dimensionality reduction
- context awareness


# Word2Vec with Gensim

**Word2Vec** is a popular technique for natural language processing (NLP) that represents words as vectors in a continuous vector space. Developed by Tomas Mikolov and his team at Google, **Word2Vec** captures semantic relationships between words based on their context within a corpus.


**Gensim** (which stands for "Generate Similar") , a robust Python library for topic modeling and document similarity, provides an efficient implementation of Word2Vec, making it accessible for both beginners and experts in the field of NLP.

# Training a Word2Vec Model

Important Parameters:

vector_size:
Number of dimensions in each word vector

window:
Context window size

min_count:
Minimum frequency of a word

workers:
Number of CPU threads  

```python
model = Word2Vec(
    sentences,
    vector_size=50,
    window=3,
    min_count=1,
    workers=4
)

print("Model trained successfully!")
```

# Vocabulary Learned by the Model

The model creates vectors for all words in the training data.

```python
list(model.wv.index_to_key)
```

# Viewing Word Vectors

Each word is represented as a numerical vector.

```python
vector = model.wv['machine']

print(vector)
```

# Finding Similar Words

Word2Vec can identify words that appear in similar contexts.

```python
model.wv.most_similar("learning")
```

# Calculating Similarity Between Two Words

Similarity values range from:

- -1 → Completely opposite
- 0  → Unrelated
- 1  → Very similar

```python
similarity = model.wv.similarity(
    "machine",
    "learning"
)

print(similarity)
```

# Word Analogy Concept

A famous property of Word2Vec is vector arithmetic.

Example:

King - Man + Woman ≈ Queen

This emerges automatically when training on very large datasets.

Our small dataset is too small to demonstrate this accurately, but the concept is one of the major strengths of Word2Vec.

# Word2Vec Architectures

Word2Vec has two architectures:

1. CBOW (Continuous Bag of Words)
2. Skip-Gram

CBOW:
Predicts a target word using surrounding words.

Example:

"The cat ___ on the mat"

Context:
The, cat, on, the, mat

Target:
sat

------------------------------------------------

Skip-Gram:
Predicts surrounding words using a target word.

Target:
sat

Predicted:
cat, on, mat

------------------------------------------------

CBOW:
Faster training

Skip-Gram:
Usually performs better on smaller datasets and rare words.

# Train CBOW Model

```python
from gensim.models import Word2Vec

cbow_model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=1,
    sg=0
)

print("CBOW model trained")
```

# Train Skip-Gram Model

```python
skipgram_model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=1,
    sg=1
)

print("Skip-Gram model trained")
```

# Compare Results

### CBOW

```python
cbow_model.wv.most_similar("learning")
```

### Skip-Gram

```python
skipgram_model.wv.most_similar("learning")
```

# Advantages of Word2Vec

1. Dense vector representation
2. Captures semantic relationships
3. Efficient training
4. Useful for many NLP tasks
5. Better than One-Hot Encoding for semantic understanding

# Limitations of Word2Vec

1. One vector per word
2. Cannot handle multiple meanings effectively

Example:

"bank" (financial institution)

"bank" (river bank)

Word2Vec gives the same vector for both meanings.

Modern models such as BERT solve this limitation using contextual embeddings.

```python
new_sentence = ["machine", "learning"]

for word in new_sentence:
    print(f"\nWord: {word}")
    print("Most Similar Words:")
    
    for similar_word, score in model.wv.most_similar(word, topn=3):
        print(similar_word, round(score, 3))
```
