# News Article Classifier : 20 Newsgroups Dataset
## Weekly Task 2 | Naive Bayes Text Classification

In this notebook we will:
- Load real news articles from sklearn
- Convert text to numbers using TF-IDF
- Train a Naive Bayes classifier
- Measure accuracy, precision, and recall
- Plot a confusion matrix

```python
# importing required libraries

from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import matplotlib.pyplot as plt
import numpy as np
```

## Step 1 : Loading the Dataset

The 20 Newsgroups dataset has thousands of real news articles.
We will only use 6 categories to keep things simple.

We load two splits:
- **train** : articles the model will learn from
- **test** : articles the model will be tested on 

```python
# We pick 6 simple categories (out of 20 available)
categories = [
    'rec.sport.hockey',
    'sci.space',
    'comp.graphics',
    'talk.politics.guns',
    'rec.autos',
    'sci.med'
]

# Load train and test data
train_data = fetch_20newsgroups(subset='train', categories=categories, remove=('headers', 'footers', 'quotes'))
test_data  = fetch_20newsgroups(subset='test',  categories=categories, remove=('headers', 'footers', 'quotes'))

print("Categories:", train_data.target_names)
print("Training articles:", len(train_data.data))
print("Testing articles: ", len(test_data.data))
```

# Peeking at the data

```python
# Let's look at one article
print("--- Sample Article ---")
print(train_data.data[0][:300])   # first 300 characters

print("\n--- Its Category ---")
print(train_data.target_names[train_data.target[0]])
```

# TF-IDF

## Step 2 : Convert Text to Numbers using TF-IDF

Computers cannot understand raw text.
We use TF-IDF  to convert
each article into a vector of numbers.

- Common words like "the", "is" get low scores
- Unique meaningful words get higher scores

```python
# TF-IDF Vectorization

vectorizer = TfidfVectorizer(max_features=3000, stop_words='english')

# Learn vocabulary from train data and convert to numbers
X_train = vectorizer.fit_transform(train_data.data)

# Convert test data using the SAME vocabulary
X_test = vectorizer.transform(test_data.data)

# Labels (correct answers)
y_train = train_data.target
y_test  = test_data.target

print("Train matrix shape:", X_train.shape)
print("Test matrix shape: ", X_test.shape)
print("Shape means: (number of articles, number of words tracked)")
```

# Naive Bayes

## Step 3 — Train the Naive Bayes Classifier

Naive Bayes learns which words appear most in each category.
For example:
- Words like "puck", "goal", "NHL" → hockey
- Words like "NASA", "orbit", "rocket" → space

Training is very fast for Naive Bayes — it just counts word patterns.

```python
# Train the model

model = MultinomialNB()
model.fit(X_train, y_train)

print("Model trained.")
```

# Evaluation

## Step 4 — Evaluate the Model

We now use the test articles (which the model has never seen)
to check how well it learned.

- **Accuracy** — overall % of correct predictions
- **Precision** — when it predicts a category, how often is it right?
- **Recall** — of all articles in a category, how many did it find?

# prediction and accuracy

```python
# Make predictions on test data
predictions = model.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {round(accuracy * 100, 3 )}%")
```

 # Precision and Recall

```python
# Full report with precision and recall for each category
report = classification_report(
    y_test,
    predictions,
    target_names=train_data.target_names
)
print(report)
```

# confusion MAtrix 

## Step 5 — Confusion Matrix

A confusion matrix shows exactly where the model made mistakes.

- The diagonal = correct predictions
- Off-diagonal = where it got confused

For example, if "sci.space" and "sci.med" share many science words,
the model might sometimes mix them up.

# Plot confusion matrix

```python
cm = confusion_matrix(y_test, predictions)
labels = train_data.target_names

plt.figure(figsize=(8, 6))
plt.imshow(cm, cmap='Blues')
plt.colorbar(label='Number of articles')

plt.xticks(range(len(labels)), labels, rotation=45, ha='right')
plt.yticks(range(len(labels)), labels)

plt.xlabel("Predicted Category")
plt.ylabel("Actual Category")
plt.title("Confusion Matrix")

# Write the count numbers inside each box
for i in range(len(labels)):
    for j in range(len(labels)):
        plt.text(j, i, cm[i][j], ha='center', va='center', fontsize=9)

plt.tight_layout()
plt.show()
```
