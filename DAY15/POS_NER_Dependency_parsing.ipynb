```python
# Importing spaCy and loading the English model

import spacy

nlp = spacy.load("en_core_web_sm")
```

# Creating a Sample Document

A document is created by passing text into the spaCy language model.

The model processes the text and stores linguistic information that can be accessed later.

```python
text = "NLP is a fascinating field. It allows us to understand and process human language. I am excited to learn more about it!"

doc = nlp(text)

print(doc)
```

# Tokenization

Tokenization is the process of breaking text into smaller units called tokens.

```python
for token in doc:
    print(token.text)
```

# Part-of-Speech (POS) Tagging

POS Tagging identifies the grammatical role of each word in a sentence.

Examples:

- Noun
- Verb
- Adjective
- Pronoun
- Adverb
- Determiner

This helps computers understand sentence structure.

```python
for token in doc:
    print(token.text, "-------->", token.pos_)
```

# Detailed POS Information

spaCy provides:

- token.text : actual word
- token.pos_ : general POS tag
- token.tag_ : detailed POS tag

```python
for token in doc:
    print(
        f"Word: {token.text:15}",
        f"POS: {token.pos_:10}",
        f"Detailed Tag: {token.tag_}",
    )
```

```python
# POS as Table

import pandas as pd

data = []

for token in doc:
    data.append([
        token.text,
        token.pos_,
        token.tag_
    ])

df = pd.DataFrame(
    data,
    columns=["Word", "POS", "Detailed Tag"]
)

df
```

# Named Entity Recognition (NER)

Named Entity Recognition identifies important real-world entities.

Examples:

PERSON:
- Elon Musk
- Lionel Messi

ORG:
- Google
- Microsoft

GPE:
- Nepal
- London

DATE:
- January 2025

NER helps in information extraction from text.

```python
text = """
Microsoft hired Jonas  in New York on Monday.
"""

doc = nlp(text)

for ent in doc.ents:
    print(ent.text, "---->", ent.label_)
```

# Displaying Entities in Table Format

```python
entity_data = []

for ent in doc.ents:
    entity_data.append([
        ent.text,
        ent.label_
    ])

entity_df = pd.DataFrame(
    entity_data,
    columns=["Entity", "Label"]
)

entity_df
```

### Visualizing NER

```python
from spacy import displacy

displacy.render(doc,style="ent", jupyter=True)
```

# Dependency Parsing

Dependency Parsing identifies relationships between words.

It answers questions such as:

- Who performed the action?
- What was the action?
- What received the action?

Example:

"The cat chased the mouse."

Subject:
cat

Verb:
chased

Object:
mouse

```python
text = "The cat chased the mouse."

doc = nlp(text)

for token in doc:
    print(
        token.text,
        token.dep_,
        token.head.text
    )
```

# Understanding Dependency Relationships

```python
dependency_data = []

for token in doc:
    dependency_data.append([
        token.text,
        token.dep_,
        token.head.text
    ])

dependency_df = pd.DataFrame(
    dependency_data,
    columns=["Word", "Dependency", "Head Word"]
)

dependency_df
```

### Dependency visualization

```python
displacy.render(doc, style="dep", jupyter=True)
```

# Combined Example

Now we will apply POS Tagging, NER, and Dependency Parsing on a single sentence.

```python
text = """
Google CEO Sundar Pichai visited London in June 2025.
"""

doc = nlp(text)

print("Sentence:")
print(text)
```

```python
print("POS TAGGING")

for token in doc:
    print(token.text, token.pos_)
```

```python
print("NAMED ENTITIES")

for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
print("DEPENDENCY PARSING")

for token in doc:
    print(
        token.text,
        token.dep_,
        token.head.text
    )
```

## complete analysis table

```python
analysis = []

for token in doc:
    analysis.append([
        token.text,
        token.pos_,
        token.dep_,
        token.head.text
    ])

analysis_df = pd.DataFrame(
    analysis,
    columns=[
        "Word",
        "POS",
        "Dependency",
        "Head Word"
    ]
)

analysis_df
```
