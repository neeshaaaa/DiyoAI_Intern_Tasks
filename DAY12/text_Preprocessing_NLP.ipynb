# What are Stopwords?
---> Stopwords are very common words that often do not add
significant meaning to NLP tasks.

Examples:
- the
- is
- are
- in
- on
- and

Removing stopwords helps:
1. Reduce noise
2. Reduce dimensionality
3. Improve efficiency


```python
# view  english stopwords

stop_words = stopwords.words('english')

print("Number of stopwords:", len(stop_words))

print("\nSome stopwords:\n")
print(stop_words[:50])
```

```python
# Remove stopwords from a sample text

text = "Stopwords are common, high-frequency words that are often filtered out during text preprocessing in Natural Language Processing (NLP) and search engine algorithms. Because they carry little semantic meaning on their own, removing them speeds up processing and focuses analysis on important keywords."

tokens = word_tokenize(text)

filtered_words = []

for word in tokens:
    
    if word.lower() not in stop_words:
        filtered_words.append(word)

print("Original Tokens:\n")
print(tokens)

print("\nAfter Stopword Removal:\n")
print(filtered_words)
```

```python
# stopword removal using list comprehension

text = "Natural Language Processing is very interesting and powerful."

tokens = word_tokenize(text)

filtered = [
    word for word in tokens
    if word.lower() not in stop_words
]

print(filtered)
```

# What is Regex?
---> Regular Expressions (Regex) are symbolic notations used in Natural Language Processing (NLP) to define specific search patterns within text.

Regex = Pattern matching technique.

Used for:
- removing punctuation
- removing numbers
- extracting emails
- extracting hashtags
- extracting URLs
- cleaning text

Regex is VERY important in NLP preprocessing.




```python
# basic regex example

text= "My phone number is 980000000 "

numbers = re.findall(r'\d+', text)

print(numbers)

# \d+ means:
# \d -> digit
# +  -> one or more
```

```python
# Remove Punctuation

text = "Hello!!! NLP is amazing, isn't it?"

clean_text = re.sub(r'[^\w\s]', '', text)

print("Original Text:")
print(text)

print("\nCleaned Text:")
print(clean_text)
```

```python
# remove numbers 

text = "My dog, bella, was born on 2020. she is 6 years old."

clean_text = re.sub(r'\d+', '', text)

print(clean_text)
```

```python
# remove extra spaces

text = " NLP             is      so  much fun! "

clean_text = re.sub(r'\s+', ' ', text)
print(clean_text)
```

```python
# extracting email addresses

text = """
Contact us:
bneesha@gmail.com
"""

emails = re.findall(r'\S+@\S+', text)

print(emails)
```

```python
# extract hastags

text= "today is am learning about #NLP and #textpreprocessing"
hashtags = re.findall(r'#\w+', text)
print(hashtags)
```

```python
# extracting urls

text = """
Visit our website: https://www.diyo.ai/ for more information.

my github: https://github.com/
"""

urls = re.findall(r'https?://\S+', text)
print(urls)
```

# text cleaning Pipeline with regex

```python
text = """
Hello!!! My name is Neesha Basnet.
I  am studying NLP since 2025.
Contact me at     bneesha@gmail.com
"""

# Lowercase
text = text.lower()

# Remove emails
text = re.sub(r'\S+@\S+', '', text)

# Remove punctuation
text = re.sub(r'[^\w\s]', '', text)

# Remove numbers
text = re.sub(r'\d+', '', text)

# Remove extra spaces
text = re.sub(r'\s+', ' ', text)

print(text)
```

# what is Stemming?

---> Stemming reduces words to root forms mechanically.

Examples:
- running -> run
- studies -> studi  
- playing -> play

Stemmed words may not always be real words.


```python
# Porter stemmer

porter = PorterStemmer()

words = [
    "running",
    "studies",
    "playing",
    "easily",
    "fairly"
]

for word in words:
    
    stem = porter.stem(word)
    
    print(f"{word} ---> {stem}")
```

```python
# lancaster stemmer

lancaster = LancasterStemmer()

words = [
    "running",
    "studies",
    "playing",
    "easily"
]

for word in words:
    
    stem = lancaster.stem(word)
    
    print(f"{word} ---> {stem}")
```

```python
# snowball stemmer

snowball = SnowballStemmer("english")

words = [
    "running",
    "studies",
    "playing",
    "beautiful"
]

for word in words:
    
    stem = snowball.stem(word)
    
    print(f"{word} ---> {stem}")
```

```python
# comapre stemmers

word = "studying"

print("Original Word:", word)

print("\nPorter Stemmer:")
print(porter.stem(word))

print("\nLancaster Stemmer:")
print(lancaster.stem(word))

print("\nSnowball Stemmer:")
print(snowball.stem(word))
```

# what is lemmatization?

--->Lemmatization reduces words to dictionary base forms.

Examples:
- running -> run
- better -> good
- cars -> car

More accurate than stemming.

```python
# basic lemmatization

lemmatizer = WordNetLemmatizer()

words = [
    "running",
    "studies",
    "playing",
    "easily",
    "fairly",
    "better",
    "cars"
]

for word in words:
    lemma = lemmatizer.lemmatize(word)
    
    print(f"{word} ---> {lemma}")
```

```python
# verb lemmatization

words = [
    "running",
    "eating",
    "driving",
    "Playing"
]

for word in words:
    
    lemma = lemmatizer.lemmatize(word, pos='v')
    
    print(f"{word} ---> {lemma}")
```

# why POS matters in lemmatization

```python
word = "better"

print("Without POS:")
print(lemmatizer.lemmatize(word))

print("\nWith adjective POS:")
print(lemmatizer.lemmatize(word, pos='a'))
```

# stemming vs lemmatization

```python
words = [
    "running",
    "studies",
    "better",
    "flying"
]

print(f"{'WORD':15} {'STEM':15} {'LEMMA':15}")

for word in words:
    
    stem = porter.stem(word)
    
    lemma = lemmatizer.lemmatize(word, pos='v')
    
    print(f"{word:15} {stem:15} {lemma:15}")
```

Lemmatization is generally better than stemming because it produces meaningful, dictionary-valid base words (lemmas) rather than arbitrarily chopped-off roots. While stemming is faster, lemmatization considers context and grammar, making it vastly superior for language understanding.



| Stemming                 | Lemmatization        |
| ------------------------ | -------------------- |
| Faster                   | Slower               |
| Rule-based cutting       | Dictionary-based     |
| Less accurate            | More accurate        |
| May create invalid words | Produces valid words |


# Full NLP Preprocessing Pipeline

```python
text = """
Hello!!! My name is Neesha Basnet.
I am studying NLP since 2025.
It is very interesting and exciting.
"""

# Lowercase
text = text.lower()

# Remove punctuation
text = re.sub(r'[^\w\s]', '', text)

# Remove numbers
text = re.sub(r'\d+', '', text)

# Tokenize
tokens = word_tokenize(text)

# Remove stopwords
tokens = [
    word for word in tokens
    if word not in stop_words
]

# Lemmatize
processed = [
    lemmatizer.lemmatize(word, pos='v')
    for word in tokens
]

print("Processed Tokens:\n")
print(processed)
```
