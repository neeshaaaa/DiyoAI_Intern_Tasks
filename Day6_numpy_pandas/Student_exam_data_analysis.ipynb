```python
# Importing libraries

import numpy as np
import pandas as pd
```

```python
# Creating Sample Student Data
student_data = {
    'Name': ['Neesha', 'Sristi', 'Riya', 'Pinki', 'Nilam', 'Niraj', 'Dhan', 'Sita', 'Gita', 'Ramesh'],
    "Math": [78, 90, 67, 88, 76, 67, 80, 85, 92, 70],
    "Science": [85, 95, 70, 92, 80, 85, 90, 95, 98, 82],
    "English": [80, 89, 72, 91, 75, 82, 88, 90, 94, 78]
}
df = pd.DataFrame(student_data)
```

```python
# Save Data to CSV

df.to_csv("students.csv", index=False)

print("CSV file created successfully!")
```

```python
data = pd.read_csv("students.csv")

print(data)
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
# Using NumPy Arrays

math_marks = np.array(data["Math"])

print(math_marks)
```

```python
# NumPy Operations

print("Average Math Marks:", np.mean(math_marks))
print("Highest Math Marks:", np.max(math_marks))
print("Lowest Math Marks:", np.min(math_marks))
```

```python
science_marks = np.array(data["Science"])

print(science_marks)
```

```python
print("Average Science Marks:", np.mean(science_marks))
print("Highest Science Marks:", np.max(science_marks))
print("Lowest Science Marks:", np.min(science_marks))
```

```python
# Add Total Marks Column

data["Total"] = data["Math"] + data["Science"] + data["English"]

print(data)
```

```python
# Add Percentage Column

total_subjects = 3
max_marks_per_subject = 100

data["Percentage"] = (
    data["Total"] / (total_subjects * max_marks_per_subject)
) * 100

print(data)
```

```python
# top students

top_student = data.loc[data["Total"].idxmax()]

print(top_student)
```

```python
# Filtering Students with Percentage Above 85

high_scorers = data[data["Percentage"] > 85]

print(high_scorers)
```

```python
# Sort students by total marks

sorted_data = data.sort_values(by="Total", ascending=False)

print(sorted_data)
```

```python
# Save final report to CSV

data.to_csv("student_report.csv", index=False)

print("Final report saved!")
```
